```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting vector

```python
#0,0 means staring from origin 0,0 and till 4,5
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,3,4, scale_units='xy', angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector1=plt.quiver(0,0,3,4, scale_units='xy', angles='xy',scale=1,color='b',label='Vector 1')
vector2=plt.quiver(0,0,5,4, scale_units='xy', angles='xy',scale=1,color='r',label='Vector 2')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.legend()
plt.show()
```
